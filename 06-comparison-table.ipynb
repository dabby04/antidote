## Full comparison table (CL + baselines)
The next cell prints one wide table: **test F1 per task**, **average F1**, and **BWT** when it is defined.

**BWT** is mainly about sequential methods; joint training may show **N/A** or a placeholder. Less negative BWT usually means less forgetting (see **`utils.compute_cl_metrics`** in the code).

```python
import sys, os
sys.path.append('/kaggle/input/datasets/dabiraomotoso/pi-detection-utils')
from utils import load_results
import json

combined = {} 

for path in [
    
    '/kaggle/input/datasets/dabiraomotoso/pi-detection-baselines/results_baselines.json',
    '/kaggle/input/datasets/dabiraomotoso/pi-detection-ewc-only/results_cl.json',
    '/kaggle/input/datasets/dabiraomotoso/pi-detection-replay-only/results_cl_replay.json',
    '/kaggle/input/datasets/dabiraomotoso/pi-detection-ewc-plus-replay/results_cl_ewc_replay.json',
]:
    if os.path.exists(path):
        combined.update(load_results(path))
        print(f'Loaded: {path}')
    else:
        print(f'Not found (skipping): {path}')

task_names = ['T1_LLMail', 'T2_HackAPrompt', 'T3_BIPIA']
all_exps   = ['static_joint', 'naive_sequential', 'ewc_only', 'replay_only', 'ewc_plus_replay']

print('\n' + '='*75)
print('FULL RESULTS TABLE')
print('='*75)
header = f'{"Method":<22}' + ''.join(f'{t:<18}' for t in task_names) + f'{"Avg F1":<10}{"BWT":<10}'
print(header)
print('-'*75)
for exp in all_exps:
    if exp not in combined:
        print(f'  {exp:<22} — not yet run')
        continue
    r   = combined[exp]
    row = f'{exp:<22}'
    for t in task_names:
        row += f'{r["per_task_f1"].get(t, 0.0):<18.4f}'
    row += f'{r["avg_f1"]:<10.4f}'
    if exp == 'static_joint':
        row += f'{"N/A":<10}'
    else:
        row += f'{str(r.get("cl_metrics", {}).get("BWT", "N/A")):<10}'
    print(row)
print('='*75)
print('BWT: less negative = less forgetting | N/A = joint training (no forgetting by design)')
```
